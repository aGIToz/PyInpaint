```python
#inpaint = pyinpaint.Inpaint("../data/lincoln_small.png",
#                  "../data/lincoln_small_mask.png")
#inpaint = pyinpaint.Inpaint("../data/colenia.png",
#                  "../data/colenia_mask.png",ps=13)
inpaint = pyinpaint.Inpaint("../data/colenia.png",
                  "../data/colenia_mask.png",ps=11)
#inpaint = pyinpaint.Inpaint("../data/colenia.png",
#                  "../data/colenia_mask.png",ps=15)
```

```python
#inpaint_img = inpaint(k_boundary=4, k_search=1000, k_patch=5)
inpaint_img = inpaint(k_boundary=4,k_search=1000, k_patch=3)
```

```python
from matplotlib import pyplot as plt
#n_img = toImg(test,(390,320,3))
import matplotlib.image as mpimg
n_img = inpaint_img
#mask = mpimg.imread("./online/lincoln_small_mask_new.png")
mask = mpimg.imread("../data/colenia_mask.png")
f = plt.figure(figsize=(10,10))
f.add_subplot(2,1, 1)
plt.imshow(n_img, cmap="gray")
f.add_subplot(2,1,2)
plt.imshow((n_img.T*mask.T).T, cmap="gray")
plt.show()
```
