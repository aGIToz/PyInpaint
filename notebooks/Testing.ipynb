```python
import pyinpaint
```

```python
inpaint = pyinpaint.Inpaint("../data/lincoln_small.png",
                  "../data/lincoln_small_mask.png")
```

```python
inpaint_img = inpaint()
```

```python
from matplotlib import pyplot as plt
#n_img = toImg(test,(390,320,3))
import matplotlib.image as mpimg
n_img = inpaint_img
#mask = mpimg.imread("./online/lincoln_small_mask_new.png")
mask = mpimg.imread("../data/lincoln_small_mask.png")
f = plt.figure(figsize=(15,15))
f.add_subplot(2,1, 1)
plt.imshow(n_img, cmap="gray")
f.add_subplot(2,1,2)
plt.imshow((n_img.T*mask.T).T, cmap="gray")
plt.show()
```
